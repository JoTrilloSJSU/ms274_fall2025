### Part 1: Horizonal Velocity Structure
First, examine the mean velocity on the west and east coasts of the U.S. Begin by loading in both the `EVEL` and `NVEL` fields from the `EVEL_2015_12.nc` and `VVEL_2015_12.nc` files, respectively -- these fields represent the zonal (eastward-flowing) and meridional (westward-flowing) velocity in the model. 

```python
# read in the EVEL and VVEL files to an xarray dataset
# note that these files must be in the same folder as this notebook
ds_evel = xr.open_dataset('EVEL_2015_12.nc')
ds_nvel = xr.open_dataset('NVEL_2015_12.nc')
```

```python
# make a plot of the velocity components (one for each)
# use a divergent colormap (such as seismic) to visualize the data
# for interpretation, it will likely be helpful to adjust the data
# bounds to be +/- 0.5 or even +/- 0.1

# make a figure
fig = plt.figure(figsize=(10,5))

# make a subplot for the eastward velocity at the surface
plt.subplot(1,2,1)
plt.pcolormesh(longitude, latitude, evel[0,0,:,:], vmin = -1, vmax = 1, cmap = 'seismic')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label = 'eastward velocity', orientation = 'horizontal')

# make a subplot for the northward velocity at the surface
plt.subplot(1,2,2)
plt.pcolormesh(longitude, latitude, nvel[0,0,:,:], vmin = -1, vmax = 1, cmap = 'seismic')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label = 'northward velocity', orientation = 'horizontal')
plt.show()
```

### Answers:
1. On the east coast there is a stronger downward velocity compared to the west coast, which has a a positive vertical velocity in the top 1000 meters below the surface.
2. The vertical velocities are 5 orders of magnitude smaller than the horizontal velocities. Stratification and Earth's rotation inhibit veritcal mixing making it a slower than typical horizontal velocities. Conversely, wind driven currents (influenced by Earth's rotation) force ocean currents at higher velocities.
3. If cold water rises like it does on the west coast, it can result in colder surface temperatures than one might expect for a given latitude. Warm water sinking along the coastline will help maintain warmer surface temperatures.
4. On the west coast, winds induce a equartorward current along the west coast. The coriolis force causes the current to turn to the right, creating an Ekman transport of surface waters offshore.  These surface waters are replaced by colder, nutrient rich water from deeper in the water column.
